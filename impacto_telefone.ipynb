```python
import pandas as pd

# Carregar o dataset
file_path = 'Impact_of_Mobile_Phone_on_Students_Health.csv'
students_health = pd.read_csv(file_path)

# Remover espaços extras dos nomes das colunas
students_health.columns = students_health.columns.str.strip()

# Exibir as primeiras linhas do dataset para verificação
print(students_health.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuração do estilo dos gráficos
sns.set(style="whitegrid")

# Gráfico de Distribuição de Idade
plt.figure(figsize=(10, 6))
sns.countplot(x='Age', data=students_health, palette='Set2')
plt.title('Distribuição de Idade dos Alunos')
plt.xlabel('Idade')
plt.ylabel('Contagem')
plt.xticks(rotation=45)  # Rotacionar os labels do eixo x para melhor visualização
plt.show()

# Gráfico de Distribuição de Gênero
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', data=students_health, palette='Set2')
plt.title('Distribuição de Gênero dos Alunos')
plt.xlabel('Gênero')
plt.ylabel('Contagem')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuração do estilo dos gráficos
sns.set(style="whitegrid")

def exibir_graficos(df):
    # Distribuição de Idade
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Age', data=df, palette='Set2')
    plt.title('Distribuição de Idade dos Alunos')
    plt.xlabel('Idade')
    plt.ylabel('Contagem')
    plt.xticks(rotation=45)  # Rotacionar os labels do eixo x para melhor visualização
    plt.show()

    # Distribuição de Gênero
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Gender', data=df, palette='Set2')
    plt.title('Distribuição de Gênero dos Alunos')
    plt.xlabel('Gênero')
    plt.ylabel('Contagem')
    plt.show()



    # Uso do Celular para Fins Educacionais
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Mobile phone use for education', data=df, palette='Set2')
    plt.title('Uso do Celular para Fins Educacionais')
    plt.xlabel('Frequência de Uso')
    plt.ylabel('Contagem')
    plt.xticks(rotation=45)  # Rotacionar os labels do eixo x para melhor visualização
    plt.show()

    # Impacto no Desempenho Acadêmico
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Performance impact', data=df, palette='Set2')
    plt.title('Impacto do Uso do Celular no Desempenho Acadêmico')
    plt.xlabel('Impacto')
    plt.ylabel('Contagem')
    plt.show()

    # Distrações
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Usage distraction', data=df, palette='Set2')
    plt.title('Distrações Causadas pelo Uso do Celular')
    plt.xlabel('Tipo de Distração')
    plt.ylabel('Contagem')
    plt.xticks(rotation=45)  # Rotacionar os labels do eixo x para melhor visualização
    plt.show()

    # Impacto na Atenção
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Attention span', data=df, palette='Set2')
    plt.title('Impacto do Uso do Celular na Atenção')
    plt.xlabel('Impacto na Atenção')
    plt.ylabel('Contagem')
    plt.show()

    # Consciência sobre Riscos à Saúde
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Health Risks', data=df, palette='Set2')
    plt.title('Consciência sobre Riscos à Saúde')
    plt.xlabel('Consciência')
    plt.ylabel('Contagem')
    plt.show()

    # Sintomas Experimentados
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Usage symptoms', data=df, palette='Set2')
    plt.title('Sintomas Experimentados')
    plt.xlabel('Sintomas')
    plt.ylabel('Contagem')
    plt.xticks(rotation=45)  # Rotacionar os labels do eixo x para melhor visualização
    plt.show()

    # Precauções de Saúde
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Health precautions', data=df, palette='Set2')
    plt.title('Precauções de Saúde Tomadas pelos Alunos')
    plt.xlabel('Precauções')
    plt.ylabel('Contagem')
    plt.xticks(rotation=45)  # Rotacionar os labels do eixo x para melhor visualização
    plt.show()

    # Avaliação Geral da Saúde
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Health rating', data=df, palette='Set2')
    plt.title('Avaliação Geral da Saúde dos Alunos')
    plt.xlabel('Avaliação de Saúde')
    plt.ylabel('Contagem')
    plt.show()

# Exibir os gráficos
exibir_graficos(students_health)
```
